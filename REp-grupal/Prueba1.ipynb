## 🔁 Mini guía práctica de flujo Git (en `Rep_grupal`)

### ✅ 1. Verifica que el repo esté limpio

```bash
git status
```

Deberías ver algo como:

> *nothing to commit, working tree clean*

---

### 🌱 2. Crea una nueva rama para una funcionalidad

Vamos a crear una rama para trabajar una funcionalidad, por ejemplo: `login`.

```bash
git checkout -b login-feature
```

Esto crea una nueva rama y te cambia a ella.

---

### ✍️ 3. Crea o edita un archivo

Por ejemplo:

```bash
echo "print('Interfaz de login')" > login.py
```

Verifica:

```bash
git status
```

Deberías ver que `login.py` está sin trackear (`untracked`).

---

### ➕ 4. Añade y haz commit

```bash
git add login.py
git commit -m "Añadido archivo login.py con mensaje inicial"
```

---

### 📜 5. Ver el historial

```bash
git log --oneline --graph --all
```

Esto te mostrará un gráfico de ramas y commits.

---

### 🔁 6. Volver a `main` y fusionar

```bash
git checkout main
git merge login-feature
```

Ahora el trabajo de la rama `login-feature` se incorpora en `main`.

---

### 🧹 7. Elimina la rama si ya no se necesita

```bash
git branch -d login-feature
```

---

## 📝 Extra: Añadir `.gitignore`

Si no quieres subir archivos `.pyc`, carpetas como `.vscode`, o entornos virtuales, crea un archivo `.gitignore` así:

```bash
echo "__pycache__/" >> .gitignore
echo ".vscode/" >> .gitignore
echo "env/" >> .gitignore
```

Y luego:

```bash
git add .gitignore
git commit -m "Añadido archivo .gitignore"
```
